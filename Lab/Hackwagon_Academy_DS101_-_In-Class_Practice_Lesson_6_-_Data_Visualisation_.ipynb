### Regular Expression Library
The Regular Expression Library (RegEx) is one of the most difficult libraries to master but the most useful as well. It helps you to clean sophisticated data in your dataset. 

RegEx recognises patterns in strings to filter them out. Understanding these patterns requires plenty of experience. Also, reading the documentation can be rather tricky so it is more advisable to search up the internet for any 
<p>https://docs.python.org/3/library/re.html</p>

````python
#imagine you have the following dirty data, and you want to clean it to remove special characters
data ="$9834&*^&%*^/89."

# Import library
import re

# .sub() is a function that takes in 1) what to replace 2) what to replace to 3) what to search through
# [^A-Za-z0-9]+ means all non alphanumeric symbols
data = re.sub('[^A-Za-z0-9]+', '', data)

# data will now be cleaned
print(data)
````

We're going to try and visualise United States of America's President Trump's tweets here: http://www.trumptwitterarchive.com

````python
import csv

tweets = []

with open('trumptweets.csv', encoding='utf-8') as file:
    reader = csv.reader(file, delimiter=',')
    
    next(reader)
    
    for row in reader:
        tweets.append(row)
tweets[:5]
````

````python
import matplotlib.pyplot as plt
# ------------------ prepare data below ------------------
# Cleaning the data to a list of words
tweet_list = []
for tweet in tweets:
    if 'https' not in tweet[0]:
        tweet_list.append(tweet[0])

# Change the list into a continuous string seperated by space
input_string = ""
for tweet in tweet_list:
    input_string += " " + tweet

# ------------------ prepare data above ------------------
from wordcloud import WordCloud, STOPWORDS

plt.figure(figsize=(15,8))
Tweets_Word_Cloud = WordCloud(width=800, 
                              height=400,
                              background_color="black", 
                              max_words=100,
                              relative_scaling = 1.0).generate(input_string) # input string here
plt.imshow(Tweets_Word_Cloud)
plt.axis("off")
plt.show()
````
